```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

Firstly, let's import our data

```python
data_train = pd.read_csv('internship_train.csv')
data_test = pd.read_csv('internship_hidden_test.csv')
```

```python
# get features and examples numbers

print(data_train.shape)
print(data_test.shape)
```

```python
pd.set_option('display.max_columns', None)
data_train.head()
```

As we see from the head of training data, there is no duplicate features and it seems that the data is noisy.

```python
cmap = plt.get_cmap('viridis')
m = plt.scatter(data_train.iloc[0:1000, 11], data_train.iloc[0:1000, -1], color=cmap(0.9), s=10) # plot a dependency between 11th feature and labels for first 1000 examples
plt.show()
```

```python
pd.options.display.max_columns = data_train.shape[1]
data_train.describe(include='all')
```

Descriptive statistics of training data shows us that there are 3 groups of features by range, mean and standard deviation.

```python
pd.set_option('display.max_columns', None)
data_test.head()
```

```python
data_test.describe(include='all')
```

The same we see in the test data except there is no labels.

Since we have no labels for test data, let's divide our train data into 2 subsets for training and testing. Choose not many examples because of the computational capabilities.

```python
x_train = data_train.iloc[0:10000, :-1].values
y_train = data_train.iloc[0:10000, -1].values
x_test = data_train.iloc[10001:13000, :-1].values
y_test = data_train.iloc[10001:13000, -1].values
```

Now let's standardize our dataset.

```python
min_max = preprocessing.MinMaxScaler()
x_train_scaled = min_max.fit_transform(x_train)
x_test_scaled = min_max.fit_transform(x_test)
x_train = pd.DataFrame(x_train_scaled)
x_test = pd.DataFrame(x_test_scaled)
```

Firstly try linear regression to get benchmark of RMSE for further modeling.

```python
lin = LinearRegression()
lin.fit(x_train, y_train)
y_pred_lin_train = lin.predict(x_train)
print("RMSE poly train:", np.sqrt(mean_squared_error(y_train, y_pred_lin_train)))
y_pred_lin = lin.predict(x_test)
print("RMSE linear:", np.sqrt(mean_squared_error(y_test, y_pred_lin)))
```

```python
start = time.time()
pipe = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
pipe.fit(x_train, y_train)
poly_pred = pipe.predict(x_train)
poly_pred_test = pipe.predict(x_test)
end = time.time()

print("RMSE poly train:", np.sqrt(mean_squared_error(y_train, poly_pred)))
print("RMSE poly test:", np.sqrt(mean_squared_error(y_test, poly_pred_test)))
print("Time for executing: {}s.".format(end-start))
```

I have chosen 2nd degree polynomial because of drastic model complexity increasing with 3rd and higher level polynomials:

```python
start = time.time()
pipe = make_pipeline(PolynomialFeatures(degree=3), LinearRegression())
pipe.fit(x_train, y_train)
poly_pred = pipe.predict(x_train)
poly_pred_test = pipe.predict(x_test)
end = time.time()

print("RMSE poly train:", np.sqrt(mean_squared_error(y_train, poly_pred)))
print("RMSE poly test:", np.sqrt(mean_squared_error(y_test, poly_pred_test)))
print("Time for executing:", end-start)
```

Now let's check polynomial regression with regularization, firstly Ridge with range of Alphas:

```python
all_preds_train_ridge = [] # create arrays for alphas, train and test RMSEs
all_preds_test_ridge = []
all_alphas_ridge = []

# try different alphas for poly ridge regression

for alpha in [10**-7, 10**-5, 0.0001, 0.001, 0.01, 0.1, 0.3, 1, 1.5, 3, 5]:
    pipe = make_pipeline(PolynomialFeatures(degree=2), Ridge(alpha=alpha))
    pipe.fit(x_train, y_train)
    poly_pred = pipe.predict(x_train)
    poly_pred_test = pipe.predict(x_test)
    all_preds_train_ridge.append(np.sqrt(mean_squared_error(y_train, poly_pred)))
    all_preds_test_ridge.append(np.sqrt(mean_squared_error(y_test, poly_pred_test)))
    all_alphas_ridge.append(alpha)
    print("Alpha: {}, RMSE ridge train: {}, RMSE ridge test: {}".format(alpha, np.sqrt(mean_squared_error(y_train, poly_pred)), np.sqrt(mean_squared_error(y_test, poly_pred_test))))
```

```python
plt.figure()
plt.plot(all_alphas_ridge, all_preds_train_ridge, color='c', label='Train RMSE')
plt.plot(all_alphas_ridge, all_preds_test_ridge, color='r', label='Test RMSE')
plt.legend()
plt.xlabel('Alpha')
plt.xscale('log')
plt.show()
```

As above, now try Lasso regularization with range of Alphas:

```python
all_preds_train_lasso = [] # create arrays for alphas, train and test RMSEs
all_preds_test_lasso = []
all_alphas_lasso = []

# try different alphas for poly lasso regression

for alpha in [10**-6, 10**-5, 0.0001, 0.001, 0.01, 0.1, 0.3, 0.5, 0.7, 0.9]: 
    pipe = make_pipeline(PolynomialFeatures(degree=2), Lasso(alpha=alpha, tol=0.02))
    pipe.fit(x_train, y_train)
    poly_pred = pipe.predict(x_train)
    poly_pred_test = pipe.predict(x_test)
    all_preds_test_lasso.append(np.sqrt(mean_squared_error(y_test, poly_pred_test)))
    all_preds_train_lasso.append(np.sqrt(mean_squared_error(y_train, poly_pred)))
    all_alphas_lasso.append(alpha)
    print("Alpha: {}, RMSE lasso train: {}, RMSE lasso test: {}".format(alpha, np.sqrt(mean_squared_error(y_train, poly_pred)), np.sqrt(mean_squared_error(y_test, poly_pred_test))))
```

```python
plt.figure()
plt.plot(all_alphas_lasso, all_preds_train_lasso, color='c', label='Train RMSE')
plt.plot(all_alphas_lasso, all_preds_test_lasso, color='r', label='Test RMSE')
plt.legend()
plt.xlabel('Alpha')
plt.xscale('log')
plt.show()
```

An analysis has shown that both Polynomial regression and Polynomial Ridge regression with small alpha show similar test RMSE but I would prefer to use Ridge because of regularization parameter that decreases parameters weight and protects us from overfitting on the actual test data.

```python
# write our data again, now using actual test set
x_train = data_train.iloc[:, :-1].values
y_train = data_train.iloc[:, -1].values
x_test = data_test.values
```

```python
# standardize data
min_max = preprocessing.MinMaxScaler()
x_train_scaled = min_max.fit_transform(x_train)
x_test_scaled = min_max.fit_transform(x_test)
x_train = pd.DataFrame(x_train_scaled)
x_test = pd.DataFrame(x_test_scaled)
```

```python
# use polynomial ridge regression to make prediction
pipe = make_pipeline(PolynomialFeatures(degree=2), Ridge(alpha=10**-7))
pipe.fit(x_train, y_train)
poly_pred = pipe.predict(x_train)
poly_pred_test = pipe.predict(x_test)
```

```python
# write prediction to a file
final_pred = open("Prediction.csv", "x")
np.savetxt(final_pred, poly_pred_test)

final_pred.close()
```
